```
data = data.astype('category')
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in data.columns:
  data[column]=le.fit_transform(data[column])
```

```
data.head()
```

**STORE FEATURE MATRIX IN X AND RESPONSE(TRAGET) IN VECTOR Y**

```
x = data.drop('class',axis=1)
y = data['class']
```

```
x
```

```
y
```

**APPLYING PCA(PRINCIPAL COMPONENT ANALYSIS)**

```
from sklearn.decomposition import PCA
pca1=PCA(n_components=7)
pcafit=pca1.fit_transform(x)
```

```
pca1.explained_variance_ratio_
```

```
sum(pca1.explained_variance_ratio_)
```

**SPLITTING DATASET INTO TRAINING SET AND TESTING SET**

```
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(pcafit,y,test_size=0.20,random_state=42)
```

```
xtrain
```

```
ytrain
```

**IMPORT MODELS**

```
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

**MODEL TRAINING**

```
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
knn = KNeighborsClassifier()
knn.fit(xtrain,ytrain)
svc = SVC()
svc.fit(xtrain,ytrain)
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
rm = RandomForestClassifier()
rm.fit(xtrain,ytrain)
bb = GradientBoostingClassifier()
bb.fit(xtrain,ytrain)
```

**PREDICTION ON TEST DATA**

```
ypred1=lr.predict(xtest)
ypred2=knn.predict(xtest)
ypred3=svc.predict(xtest)
ypred4=dt.predict(xtest)
ypred5=rm.predict(xtest)
ypred6=bb.predict(xtest)
```

**EVALUATING**

```
from sklearn.metrics import accuracy_score
```

```
print("acc lr",accuracy_score(ytest,ypred1))
print("acc knn",accuracy_score(ytest,ypred2))
print("acc svc",accuracy_score(ytest,ypred3))
print("acc dt",accuracy_score(ytest,ypred4))
print("acc rm",accuracy_score(ytest,ypred5))
print("acc bb",accuracy_score(ytest,ypred6))
```

```
fdata = pd.DataFrame({'models':['lr','knn','svc','dt','rm','bb'],
              'acc':[accuracy_score(ytest,ypred1)*100,
              accuracy_score(ytest,ypred2)*100,
              accuracy_score(ytest,ypred3)*100,
              accuracy_score(ytest,ypred4)*100,
              accuracy_score(ytest,ypred5)*100,
              accuracy_score(ytest,ypred6)*100]})
```

```
fdata
```
